##     Scipy:  

     We  have  the  min and  max temperatures  in a city  In  India for each  months of the year.  
     We  would  like  to  find  a  function  to  describe  this  and  show  it  graphically,  the  dataset  
     given below.  

     Task:  

     1. fitting it to the periodic function  

     2. plot the fit  

     Data  

     Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25  

     Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18 

```python
import matplotlib.pyplot as plt
%matplotlib inline
months=list(range(1,13))                                     # generating 1 to 12 number for Months
Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25] 
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18] 

plt.plot(months,Max,'ro')                                    # Ploting the Maximum temperature in red color
plt.plot(months,Min,'bo')                                    # Ploting the Minimum temperature in blue color
plt.xlabel("Months")
plt.ylabel("Temperature")
```

```python
from scipy import optimize
import numpy as np
def derviedfunc(times, avg, ampl, time_offset):                                    # Defining the curvefit function
    return (avg
            + ampl * np.sin((times + time_offset) * 2 * np.pi / times.max()))      # Generating the cosine function

res_max, cov_max = optimize.curve_fit(derviedfunc, months,Max,[40,20,0])           # Optimising and generating th curve fit 
res_min, cov_min = optimize.curve_fit(derviedfunc, months,Min,[-40,20,0])
monthsVal = np.linspace(0, 12, num=365)                                            # Generating the 365 points to indicate days

plt.figure()
plt.plot(months, Max, 'ro')
plt.plot(monthsVal, derviedfunc(monthsVal, *res_max), 'r-')                        # Ploting the curve fit function for Max temp
plt.plot(months, Min, 'bo')
plt.plot(monthsVal, derviedfunc(monthsVal, *res_min), 'b-')                        # Ploting the curve fit function for Min temp
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
